```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
data=fetch_california_housing()
data
```

```python
x=data.data
x
```

```python
y=data.target
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
m_model=LinearRegression()
```

```python
m_model.fit(x_train,y_train)
y_predict=m_model.predict(x_test)
y_predict
```

```python
r2=r2_score(y_test,y_predict)
r2
```

```python
mae=mean_absolute_error(y_test,y_predict)
mae
```

```python
mse=mean_squared_error(y_test,y_predict)
mse
```

```python
import numpy as np
rmse=np.sqrt(mse)
rmse
```

```python
y_train_line=m_model.predict(x_test)
```

```python
plt.scatter(y_test,y_train_line, color='blue', label="Predicted vs Actual")

# Perfect Fit Line (based on training data range)
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    color='red', linestyle='--', label="Perfect Fit Line"
)
plt.show()
```
