```python
# check before scaling the values
Q1 = df['total_spent'].quantile(0.25)
Q3 = df['total_spent'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['total_spent'] < lower_bound) | (df['total_spent'] > upper_bound)]
print("Outliers in 'total_spent' column:")
display(outliers)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['total_spent'], kde=True)
plt.title('Distribution of Total Spent')
plt.xlabel('Total Spent')
plt.ylabel('Frequency')
plt.show()
```

```python
median_total_spent = df['total_spent'].median()

df['total_spent'] = np.where(
    (df['total_spent'] < lower_bound) | (df['total_spent'] > upper_bound),
    median_total_spent,
    df['total_spent']
)

print("Outliers in 'total_spent' column replaced with median.")
```
