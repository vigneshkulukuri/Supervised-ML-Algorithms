KNN (K-Nearest Neighbors)


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.metrics import confusion_matrix,accuracy_score,roc_auc_score,log_loss
from sklearn.model_selection import train_test_split
```

```python
data=load_iris()
data
```

```python
x=data.data
y=data.target
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=12)
```

```python
knn=KNeighborsClassifier(n_neighbors=21,metric='minkowski',p=1)
```

```python
knn.fit(x_train,y_train)
y_predict=knn.predict(x_test)
y_probality_predict=knn.predict_proba(x_test)
y_predict
```

```python
confusion=confusion_matrix(y_test,y_predict)
confusion
```

```python
acc=accuracy_score(y_test,y_predict)
acc
```

```python
loss=log_loss(y_test,y_probality_predict)
loss
```

```python
roc_auc=roc_auc_score(y_test,y_probality_predict,multi_class='ovr')
roc_auc
```
